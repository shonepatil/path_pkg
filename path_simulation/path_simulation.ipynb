```python
import numpy as np
import control
from control.matlab import *  # MATLAB-like functions
from matplotlib import pyplot as plt

def plot_path_ecg(car_pose, point1, point2):
    x0 = [2, 2, 2, 2]
    x01 = [1, 1, 2, 2]
    x0 = np.array(x0)
    test = np.array([x01]).transpose()

    xhat = x0
    print(x0)
    print(test.transpose())
    xhat_mat = np.zeros([4, 5], dtype=float)
    xhat_mat[:, 0] = test.transpose()
    print(xhat_mat)

    A = np.array([[1, 1, 1, 1],
                  [1, 0, 1, 0],
                  [9, 10, 11, 12],
                  [5, 14, 15, 2]])
    LB = np.array([[0],
                   [253.96825397],
                   [0.],
                   [203.47476915]])
    a_dot_x = np.dot(A, xhat).reshape(4,1)
    # a_dot_x = a_dot_x.reshape(4,1)
    print(f"A: {A}")
    print(f"x: {xhat}")
    print(f"a dot x: {np.dot(A, xhat)}")
    print(f"a dot x: {a_dot_x}")
    print(f"a dot x + b: {np.add(a_dot_x, LB)}")
    # print(np.dot(np.dot(A, xhat), np.dot(A, xhat)))
    # print(A * x0)


    # print(x0)
    u_num_rows = x0.size
    # print(x0.size)
    # print(u_num_rows)
    x1,x2 = np.partition(A[3], 1)[0:2]
    print(f"x1,x2: ({x1},{x2})")
    efa_mag1_index = np.where(A[3] == x1)
    efa_mag1_index = np.argwhere(A[3] == x1)[0][0]
    print(f"index in A: {efa_mag1_index}")

    xs = np.array(x0) - 5
    print(xs)
    print(type(xs))

    def fun_func():
        return 1, 2

    x1,x2 = fun_func()
    print(x1, x2)

    print(np.sign(-1))
    line_error_threshold = 0.001
    x_path = linspace(0, 20, 10)
    x_car = linspace(0, 20, 10)

    P1y = point1[1]
    P2y = point2[1]
    P1x = point1[0]
    P2x = point2[0]
    delta_x = P2x - P1x + line_error_threshold
    delta_y = P2y - P1y + line_error_threshold

    y_c = car_pose[1]
    x_c = car_pose[0]
    m_path = delta_y / delta_x
    b_path = P1y - m_path * P1x
    y_path = m_path * x_path + b_path

    m_car = -1 / m_path
    b_car = y_c - m_car * x_c
    y_car = m_car * x_car + b_car

    ecg_x = (b_path - b_car) / (m_car - m_path)
    ecg_y = m_car * ecg_x + b_car

    ecg_r = np.power((np.power(ecg_x-x_c,2) + np.power(ecg_y-y_c,2)),0.5)
    e_cg_sign = np.sign(m_car)
    e_cg = float(e_cg_sign * ecg_r)
    
    print("m_path: {}, m_car: {}".format(m_path, m_car))
    print(f"ecg_r: {ecg_r}, e_cg_sign: {e_cg_sign}, e_cg: {e_cg}")
    x_car_min = np.min([ecg_x, x_c])
    x_car_max = np.max([ecg_x, x_c])

    fig, ax = plt.subplots()
    ax.scatter(x_c, y_c, label="car pose")
    ax.scatter([P1x, P2x], [P1y, P2y], label="path points")
    ax.scatter(ecg_x, ecg_y, label="e_cg point")
    ax.plot(x_path, y_path, label="path")
    ax.plot(x_car, y_car, label="e_cg line")
    ax.axis('equal')
    ax.set_xlim(0, 12)
    ax.set_ylim(0, 7)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    
    
    plt.plot(temp['px'].values, temp['py'].values)
#     plt.title('Trajectory Recorded')
#     plt.xlabel('x')
#     plt.ylabel('y')
    plt.plot(9.735, 0, '-ro', markersize=5)
#     plt.show()
    
    plt.show()
```
